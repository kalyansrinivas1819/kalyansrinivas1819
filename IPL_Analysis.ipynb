```python
def runs_scored(batsman_name):
    vk=deli[deli['batsman']=='batsman_name']
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(2)
```

```python
runs_scored('MS Dhoni')
```
